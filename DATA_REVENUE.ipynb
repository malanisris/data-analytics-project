```python
# drop null values
df.dropna(inplace=True)
```

```python
df = df.rename(columns={'RECIEVER': 'reciever'})

print(df)
```

```python
# change data type
df['AMOUNT'] = df['AMOUNT'].astype('int')
df['BILL NO.']=df['BILL NO.'].astype('int')
df['Revenue_per_Quintal']=df['Revenue_per_Quintal'].astype('int')
```

# Data Analysis

```python
# plotting a bar chart for AMOUNT and it's count

ax = sns.countplot(x = 'AMOUNT',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
CITY = df.groupby(['AMOUNT'], as_index=False)['CITY'].sum().sort_values(by='CITY', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = CITY, x = 'AMOUNT',y= 'CITY')
```

```python
# top 10 most sold products (same thing as above)

fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('DATE')['AMOUNT'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```

```python
#  Step 4: SQL Integration with SQLite
import sqlite3

conn = sqlite3.connect("project_data.db")
df.to_sql("transactions", conn, if_exists="replace", index=False)

# Query: Monthly Sales
query = '''
SELECT MONTH, SUM(AMOUNT) AS total_sales
FROM transactions
GROUP BY MONTH
ORDER BY total_sales DESC;
'''
pd.read_sql_query(query, conn)
```
